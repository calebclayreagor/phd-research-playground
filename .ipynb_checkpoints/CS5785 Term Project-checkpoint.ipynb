```python
import tigramite
from tigramite import data_processing as pp
from tigramite import plotting as tp
from tigramite.pcmci import PCMCI
from tigramite.independence_tests import ParCorr, GPDC, CMIknn, CMIsymb
```

```python
data = s.expression[0]
var_names = data.columns

dataframe = pp.DataFrame(data.values, datatime = s.t[0].values, var_names=var_names)
```

```python
parcorr = ParCorr(significance='analytic')

pcmci = PCMCI(dataframe=dataframe, cond_ind_test=parcorr, verbosity=2)

results = pcmci.run_pcmci(tau_max=1, pc_alpha=None)

q_matrix = pcmci.get_corrected_pvalues(p_matrix=results['p_matrix'], fdr_method='fdr_bh')
```

```python
pcmci.print_significant_links(p_matrix = results['p_matrix'], q_matrix = q_matrix, 
                              val_matrix = results['val_matrix'], alpha_level = 0.01)
```
