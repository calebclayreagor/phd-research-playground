## Theoretical model

```python
model = branching_model(f'{wd}/3dpf/Composite_3.tiff', xy_res = 65., z_res = 200.) 
```

```python
model.segment_stack(diameter=60)
```

```python
model.show_segmentation(separate_channels=False, annotate=[[530,570],[25]])
```

```python
model.simulate_metropolisMC(start=[[530,570],[25]], nsteps=500, nseeds=1000, d=1.0, σ_signal=2.0, U_membrane=1000)
```

```python
# ISSUES: Axons are not points... they have volume (area in 2d)... probably need to account for this
#        - Fewer (and probably longer) steps (some axons go away from cells, likely in one or two "steps")
#        - How to measure if real data drawn from simulated distribution?

model.plot_trajectories()
```

```python
from skimage.draw import ellipse_perimeter, ellipse

xx, yy = ellipse_perimeter(520, 65, 250, 35)
xx_, yy_ = ellipse(520, 65, 250, 35)
ellipse_mask = np.zeros(model.xz_maxNorm[...,1].shape)
ellipse_mask[yy_, xx_] = 1.

fig, ax = plt.subplots(2,1, sharex=True, sharey=True)

axons = model.xz_maxNorm[...,1].copy()
trajs = model.pos_mask.copy()


axons[np.where((axons<0.2)|(model.xz_maxMask > 0.)|(ellipse_mask==0.))] = 0.
axons[axons>0.] = 1.

ax[0].imshow(axons, aspect=model.aspect, origin='lower', cmap='gray')
ax[0].scatter(xx, yy, s=1, c='r')

trajs[np.where((model.xz_maxMask > 0.)|(ellipse_mask==0.))] = 0.

ax[1].imshow(trajs, aspect=model.aspect, origin='lower', cmap='gray')
ax[1].scatter(xx, yy, s=1, c='r')


plt.show()
```
